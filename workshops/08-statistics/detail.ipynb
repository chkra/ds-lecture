### Konfidenzintervalle und Interquantilbereiche

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

std = 20
N = 15

# Ziehen Sie eine weitere Stichprobe für sample_ber
sample_ber_2 = np.random.normal(mean_ber, std, N)
plt.boxplot(sample_ber_2, notch=True, vert=False)

# Zeichnen Sie Linien durch die Mittelwerte
plt.axvline(mean_ber, color='blue', linestyle='--')

# Berechnen Sie den Mittelwert und das Konfidenzintervall der neuen Stichprobe
avg_ber_2 = np.mean(sample_ber_2)
conf_int_ber_2 = stats.t.interval(0.95, len(sample_ber_2)-1, loc=avg_ber_2, scale=stats.sem(sample_ber_2))

# Zeichnen Sie die Konfidenzintervalle
plt.axvline(conf_int_ber_2[0], color='green', linestyle=':')
plt.axvline(conf_int_ber_2[1], color='green', linestyle=':')
plt.xlim([100, 250])

# Zeigen Sie den Plot an
plt.show()
```

**Intuition:** Unsere Stichprobe hat eine ziemlich große Schwankung (es gibt sehr große und sehr kleine Männer in Berlin), das sieht man an den Outliern des Boxplots. Der Interperzentilbereich (=der dicke Teil des Bockplots) zeigt uns, wo die meisten Männer angesiedelt sind. Wir können den Durchschnitt unserer Stichprobe berechnen (rote Linie im Boxplot). Diese ist nahe an dem (geheimen) echten Mittelwert dran (blaue Linie), den wir im echten Leben aber nicht kennen. Basierend auf der Students t-Distribution und unserer Stichprobe wissen wir aber ziemlich sicher (mit 95%iger Wahrscheinlichkeit), dass der (geheime) echte  Mittelwert im Konfidenzintervall (grüne Linien) um den gemessenen Durchschnitt liegt.

**Beobachtung 1:** Je größer die (echte, geheime) Standardabweichung ist, desto unsicherer sind wir: das Konfidenzintervall wird größer.

**Beobachtung 2:** Je größer unsere Stichprobe ist, desto sicherer werden wir: das Konfidenzintervall wird kleiner.

**Anmerkung:** Manchmal werden Boxplot-artige Darstellungen verwendet, um einen Mittelwert und sein Konfidenzintervall darzustellen. Das ist okay, bedeutet aber eben etwas anderes als als ein Boxplot von einem Durchschnitt und dem Interperzentilbereich.

### Gesetz der großen Zahlen (Vereinfachung)

Angenommen, wir haben eine große Stichprobe von unabhängigen N Werten X<sub>1</sub>, ..., X<sub>N</sub>, die aus irgendeiner Verteilung mit Mittelwert μ und Varianz &sigma;<sup>2</sup> entnommen wurden. Dann wäre für ausreichend große N  der Durchschnitt $\sum_i X_i/N$ normalverteilt, mit Mittelwert μ und Varianz $\sigma^2 /N$.

**Unsere Schlussfolgerung - Intuition:**
Wir haben viele Münchner genommen - der Durschnitt zeigt vermutlich den echten Münchner Mittelwert. Wir haben viele Berliner genommen - der Durschnitt zeigt vermutlich den echten Berliner Mittelwert. Die Mittelwerte sind sehr ähnlich. Es ist _wahrscheinlich_, dass beide Stichproben aus der gleichen Verteilung kommen.

### Besuch der Niederländischen Fans

```python
std = 20
N = 5000


sample_ber = np.random.normal(mean_ber, std, N)
sample_ned = np.random.normal(mean_ned, std, N)

plt.hist(sample_ber, bins=10, alpha=0.3, label='Stichprobe Berlin', color='blue')
plt.hist(sample_ned, bins=10, alpha=0.3, label='Stichprobe Niederlande',  color='orange')

avg_ber = np.mean(sample_ber)
avg_ned = np.mean(sample_ned)

plt.axvline(avg_ber, color='blue', linestyle='--')
plt.axvline(avg_ned, color='orange', linestyle='--')

plt.legend()
plt.show()
```

Das kann doch kein Zufall sein, oder?

### Hypthesentests

```python
plt.hist(sample_ber, bins=10, alpha=0.3, label='Stichprobe Berlin', color='blue')
plt.hist(sample_ned, bins=10, alpha=0.3, label='Stichprobe Niederlande',  color='orange')
plt.axvline(avg_ber, color='blue', linestyle='--')
plt.axvline(avg_ned, color='orange', linestyle='--')
plt.legend()
plt.title("Verteilung der Werte in beiden Stichproben")
plt.show()
```

**Intuition:** Wir haben viele Berliner, sind uns also beim Berliner Mittelwert recht sicher (GdgZ), wir haben genausoviele Niederländer, sind uns also beim Mittelwert auch sicher. Die Mittelwerte scheinen unterschiedlich zu sein!

```python
std = 20 # natürliche Größenschwankung in Zentimeter
N = 5000 

sig = std / np.sqrt(N)  # Schwnakung des Mittelwertes nach Gesetz d. Großen Zahlen (beide Stichproben haben gleiche Größe N)

x = np.linspace(170, 180, 100)
y_ber = norm.pdf(x, avg_ber, sig)
y_ned = norm.pdf(x, avg_ned, sig)
plt.plot(x, y_ber, label='Mittelwert Berlin')
plt.plot(x, y_ned, label='Mittelwert Niederlande')
plt.axvline(avg_ber, color='blue', linestyle='--')
plt.axvline(avg_ned, color='orange', linestyle='--')
plt.title("Mittelwerte und Unsicherheit ihrer Schätzung bei dieser Stichprobengröße")
plt.legend()
plt.show()
```

Students t-test: sind die Mittelwerte beider Stichproben gleich?

**Intuition**: 
Wir nehmen an (Nullhypothese), dass wir den Mittelwert `avg_ned` einer Stichprobe beobachten, die aus einer Population mit dem wahren Mittelwert `avg_ber` stammt. Wie wahrscheinlich ist diese Beobachtung? 

```python
from scipy import stats

t_stat, p_val = stats.ttest_ind(sample_ber, sample_ned)

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_val}")
```

Der p-value sagt: es ist sehr, sehr unwahrscheinlich! Wir lehnen die Nullhypothese ab!

### Zum Schluss: was kann schief gehen?

## <img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/93/Beta-Fehler.png/700px-Beta-Fehler.png">

Hypthesentest: 
* Merkmale zweier Stichprobe werden verglichen
* Unsere Nullhypothese ist: die zweite Stichprobe ist "unspektakulär", das ist nur natürliche Schwankung, die wahren Mittelwerte sind identisch!
* Unter der Annahme, dass die zweite Stichprobe (mit $\mu_1$) aus einer unspektakulären Verteilung (mit $\mu_0$) kommt, wie wahrscheinlich ist die Beobachtung?
* Signifikanzwert  (sowas wie "0.05"): Wahrscheinlichkeit in der ersten Kurve, ab der wir die Nullhypothese ablehnen
* alpha-Fehler, Mittelwert $\mu_1$ ist außerhalb der Signifikanzgrenze: Nullhypothese wird zurückgewiesen, obwohl sie wahr ist (falscher Alarm): kann man vermeiden durch harte (kleine) Signifikanz-Grenzen
* beta-Fehler, Mittelwert $\mu_1$ ist innerhalb der Signifikanzgrenze: Nullhypthose wird akzeptiert, obwohl sie falsch ist (wir übersehen etwas spannendes): kann man vermeiden durch weiche (betragshohe) Signifikanzgrenzen und großes $N$ (dann werden die Glockenkurven schmaler = der overlap geringer)
